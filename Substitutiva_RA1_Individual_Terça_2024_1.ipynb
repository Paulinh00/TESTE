expandi o numerador

$$\lim_{x \to 0} \frac{(\sqrt{x+9} - 3)(\sqrt{x+9} + 3)}{x(\sqrt{x+9} + 3)} = \lim_{x \to 0} \frac{(x+9) - 9}{x(\sqrt{x+9} + 3)}$$

simplificas

$$\lim_{x \to 0} \frac{x}{x(\sqrt{x+9} + 3)} = \lim_{x \to 0} \frac{1}{\sqrt{x+9} + 3}$$

Substitiu x = 0:

$$\lim_{x \to 0} \frac{1}{\sqrt{x+9} + 3} = \frac{1}{\sqrt{0+9} + 3} = \frac{1}{3+3} = \frac{1}{6}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Definindo a variável simbólica e a função
x = sp.symbols('x')
funcao = (25*x + 7) / sp.sqrt(5*x**2 + 2)

# Encontrando as raízes da função
raizes = sp.solve(funcao, x)

# Convertendo as raízes para números decimais
raizes_decimais = [float(raiz) for raiz in raizes]

# Imprimindo as raízes encontradas
print("Raízes da função:", raizes_decimais)

# Criando o intervalo de valores de x para o gráfico
x_vals = np.linspace(0, 2, 400)

# Convertendo a expressão simbólica para uma função numérica
funcao_numeric = sp.lambdify(x, funcao, 'numpy')

# Calculando os valores da função para o intervalo dado
y_vals = funcao_numeric(x_vals)

# Plotando o gráfico da função
plt.plot(x_vals, y_vals, label='(25*x + 7) / sp.sqrt(5*x**2 + 2)')
plt.axhline(0, color='black', linewidth=0.5)  # Adicionando a linha horizontal y=0
plt.scatter(raizes_decimais, [0]*len(raizes_decimais), color='red', label='Raízes')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

$$\lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} $$

$$ = \lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2} \sqrt{1 + \frac{2}{5x^2}}} $$

Separando a raiz quadrada no denominador, obtemos:

$$\lim_{x \to \infty} \frac{\dfrac{25x + 7}{x^2}}{\dfrac{\sqrt{5x^2}}{x^2} \cdot \sqrt{1 + \frac{2}{5x^2}}} $$

Simplificando com a regra de potências

$$ = \lim_{x \to \infty} \frac{25x + 7}{5x \sqrt{1 + \frac{2}{5x^2}}} $$


$$ = \lim_{x \to \infty} \frac{5 + \frac{7}{x}}{5 \sqrt{1 + \frac{2}{5x^2}}} $$

tendem a zero. Isso porque qualquer número finito dividido por infinito se torna infinitamente pequeno (tende a zero).

Então, podemos reescrever a expressão:

$$\lim_{x \to \infty} \frac{5 + \dfrac{7}{x}}{5 \sqrt{1 + \frac{2}{5x^2}}} = \frac{5 + 0}{5 \sqrt{1 + 0}} $$

$$ = \frac{5}{5} $$

dividondo chegamos em

"infinito menos infinito" quando x se aproxima do infinito negativo. Geralmente, essa forma não possui um limite finito

$$\lim_{x \to -\infty} (\sqrt{x^2 - x} - x)$$

Usamos uma técnica chamada racionalização para eliminar a raiz quadrada do denominador. Isso é feito multiplicando o numerador e o denominador pelo conjugado do radical, que é {x^2 - x} - x

$$\lim_{x \to -\infty} \left(\frac{(\sqrt{x^2 - x} - x)(\sqrt{x^2 - x} + x)}{\sqrt{x^2 - x} + x}\right)$$

simplificamos:
Multiplicando os dois fatores no numerador e simplificando a expressão resultante, obtemos uma nova expressão que pode ser mais fácil de trabalhar quando x se aproxima do infinito negativo.

$$\lim_{x \to -\infty} \left(\frac{(x^2 - x) - x^2}{\sqrt{x^2 - x} + x}\right)$$

Simplificação Adicional
Podemos simplificar ainda mais a expressão combinando termos no numerador.

$$\lim_{x \to -\infty} \left(\frac{-x}{\sqrt{x^2 - x} + x}\right)$$

Usando Equivalências para Infinito Negativo

À medida que x se aproxima do infinito negativo, o termo -1/x
  se aproxima de zero. Podemos usar essa equivalência para reescrever a expressão.

$$\lim_{x \to -\infty} \left(\frac{-1}{\sqrt{1 - \frac{1}{x}} + 1}\right)$$

agora vem a substiuicao Embora possamos substituir o infinito negativo diretamente em algumas expressões que se aproximam de um valor finito específico quando x se aproxima do infinito positivo ou negativo, nem sempre é esse o caso. No nosso caso, substituir o infinito negativo diretamente na raiz quadrada levaria a uma forma indeterminada.

x se aproxima do infinito negativo, o termo -1/x se aproxima de zero e dai simplifica

$$\frac{-1}{\sqrt{1 + 0} + 1}$$

entao

$$-\frac{1}{2}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Definindo a variável simbólica
x = sp.symbols('x')

# Definindo a expressão da função
funcao = (4*x**2 + 3*x - 5) / (2*x**2 - x + 1)

# Calculando o limite para o ponto desejado (-2)
ponto_desejado = 0
y_marker = sp.limit(funcao, x, ponto_desejado)

# Preparando os valores de x para o gráfico, evitando divisão por zero perto de -2
x_vals = np.linspace(0, 2, 400)
x_vals = x_vals[x_vals != ponto_desejado]
y_vals = [(x_val**2 - 1 ) / (x_val - 1) for x_val in x_vals]

# Criando o gráfico
plt.figure(figsize=(10, 10))
plt.plot(x_vals, y_vals, label='f(x)')
plt.axvline(x=ponto_desejado, color='orange', linestyle='--', label=f'x = {ponto_desejado}')
plt.scatter(1,2, color="red", label="x=1")
# Convertendo y_marker para float para plotagem e ajustando linha horizontal
y_marker_float = float(y_marker)


# Adicionando título, legendas e ajustando o eixo y para melhor visualização
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
#plt.ylim(-10, 10)
plt.grid(True)
plt.show()
```

Fatorando o numerador e o denominador:

Comecamos fatorando tanto o numerador (4x^2 + 3x - 5) quanto o denominador (2x^2 - x + 1) usando técnicas de fatoracão de polinômios. No numerador, podemos fatorar um 2 dos dois primeiros termos e usar o padrão soma-produto para obter (2x - 1)(2x + 5). No denominador, podemos novamente fatorar um 2 e usar o padrão diferença de quadrados para obter (2x - 1)(x - 1).

$$\lim_{x \to \infty} \frac{4x^2 + 3x - 5}{2x^2 - x + 1} = \lim_{x \to \infty} \frac{(2x - 1)(2x + 5)}{(2x - 1)(x - 1)}$$


Cancelando o fator comum:

Como tanto o numerador quanto o denominador compartilham um fator comum de (2x - 1), podemos dividir ambos por ele. Isso é permitido desde que x ≠ 1/2 (o que não é problema aqui porque estamos tomando o limite à medida que x se aproxima do infinito). Este cancelamento simplifica a expressão.


$$\lim_{x \to \infty} \frac{(2x - 1)(2x + 5)}{(2x - 1)(x - 1)} = \lim_{x \to \infty} \frac{2x + 5}{x - 1}$$

Calculando o limite:

Agora, temos uma expressão mais simples com apenas x e constantes. À medida que x aumenta em direção ao infinito positivo (representado por x → ∞), os termos com x no numerador e denominador se tornam cada vez mais dominantes em comparacão com os termos constantes. Por exemplo, 5/x se torna muito próximo de 0 para valores muito grandes de x.

Portanto, podemos aproximar a expressão dividindo tanto o numerador quanto o denominador pela maior potência de x presente ou seka x

$$\lim_{x \to \infty} \frac{2x + 5}{x - 1} = \lim_{x \to \infty} \frac{2 + \frac{5}{x}}{1 - \frac{1}{x}}$$

à medida que x se aproxima do infinito, tanto 5/x quanto 1/x se aproximam de zero. Isso nos deixa com uma fracão simples com constantes

$$\lim_{x \to \infty} \frac{2 + \frac{5}{x}}{1 - \frac{1}{x}} = \frac{2 + 0}{1 - 0} = \frac{2}{1} = \boxed{2}$$
