## Solução Questão 1

Análise da Primeira Derivada:

A primeira derivada é utilizada para identificar a taxa de mudança instantânea da função. Ela é importante para achar pontos onde essa derivada é igual a zero, que podem ser máximos ou mínimos locais.

$$Pontos Críticos Encontrados:$$

A linha verde, que representa a primeira derivada, intercepta o eixo horizontal nos pontos $(2.33)$ e $(3.00)$. Esses são os pontos críticos.

$$Classificação dos Pontos Críticos:$$

A classificação quanto à natureza desses pontos críticos depende do sinal da segunda derivada (linha vermelha).
Um sinal negativo indica um máximo local, enquanto um sinal positivo indica um mínimo local.

$$Análise da Segunda Derivada:$$

A segunda derivada indica a concavidade da curva da função.
Ponto de Inflexão:
O ponto de inflexão ocorre em $(2.67)$, que é onde a segunda derivada intercepta o eixo horizontal.

Resumo dos Pontos de Extremo e Inflexão:

$$Máximo Local:$$

No ponto $(2.33)$, a primeira derivada é zero e a segunda derivada é negativa, indicando um máximo.

$$Mínimo Local:$$

No ponto $(3.00)$, a primeira derivada também é zero, mas a segunda derivada é positiva, indicando um mínimo.

$$Ponto de Inflexão:$$

Em $(2.67)$, há uma mudança no sinal da segunda derivada, sugerindo uma alteração na concavidade.

Para encontrar os pontos críticos, igualamos a derivada a zero:
$$4x^3 - 12x^2 + 8x = 0.$$
$$4x(x^2 - 3x + 2) = 0$$
$$x(x - 1)(x - 2) = 0$$

Os pontos críticos são $$x = 0, x = 1, x = 2$$

Para determinar a natureza destes pontos, usamos a segunda derivada:
$$f(x) = 12x^2 - 24x + 8$$
Substituindo os valores críticos:
- \( f(0) = 8 > 0 \) (mínimo local)
- \( f(1) = -4 < 0 \) (máximo local)
- \( f(2) = 8 > 0 \) (mínimo local)


$$RESOLUCAO(B)$$
A função é $$f(x) = x^4 (x - 3)^2$$

Primeiro, encontramos a primeira derivada usando a regra do produto:
$$f'(x) = 4x^3(x - 3)^2 + 2x^4(x - 3)$$
$$f'(x) = x^3(4(x-3)^2 + 2x(x-3))$$
$$f'(x) = x^3(4x^2 - 20x + 36)$$

Igualando a zero:
$$x^3(4x^2 - 20x + 36) = 0$$

Os pontos críticos são \( x = 0 \) (onde \( x^3 = 0 \)) e as raízes de \( 4x^2 - 20x + 36 = 0 \) não são reais, portanto, o único ponto crítico real é \( x = 0 \).

Analisando a segunda derivada para determinar a natureza em \( x = 0 \):
$$f''(x) = 12x^2(x-3)^2 + 16x^3(x-3) + 2x^4$$
$$f''(0) = 0$$

Como \( f''(0) = 0 \), precisaríamos da terceira derivada para análise mais profunda, indicando que \( x = 0 \) é um ponto de inflexão.

## Solução Questão 4##

1. **Distância da travessia do rio**:

   - O pescador começa em $(0, 0)$ e deseja chegar a $(x, 10)$.
   - A distância da travessia é $\sqrt{x^2 + 10^2}$.
   - A velocidade do pescador ao remar é $2 \text{ m/s}$.
   - O tempo necessário para a travessia é $\frac{\sqrt{x^2 + 10^2}}{2}$.

2. **Distância da caminhada**:

   - Após a travessia, o pescador precisa caminhar de $(x, 10)$ até $(10, 10)$.
   - A distância da caminhada é $10 - x$.
   - A velocidade do pescador ao caminhar é $4 \text{ m/s}$.
   - O tempo necessário para a caminhada é $\frac{10 - x}{4}$.

3. **Tempo Total**:

   - O tempo total $T(x)$ é a soma do tempo de travessia e do tempo de caminhada:

$$
T(x) = \frac{\sqrt{x^2 + 10^2}}{2} + \frac{|10 - x|}{4}
$$


Vamos encontrar o ponto $x$ que minimiza a função $T(x)$, que representa o tempo total que o velho pescador levaria para atravessar o rio e caminhar até a casa de seu amigo. A função é:

$$
T(x) = \frac{\sqrt{x^2 + 100}}{2} + \frac{|10 - x|}{4}
$$


Primeiramente, derivamos a função $T(x)$:

$$
\frac{d}{dx}T(x) = \frac{x}{2 \sqrt{x^2 + 100}} - \frac{10 - x}{4|10 - x|}
$$

Vamos encontrar o ponto crítico resolvendo $\frac{d}{dx}T(x) = 0$:

$$
\frac{x}{2 \sqrt{x^2 + 100}} = \frac{10 - x}{4|10 - x|}
$$


$x \leq 10$

$$
\frac{x}{2 \sqrt{x^2 + 100}} = \frac{10 - x}{4 \cdot (10 - x)}
$$

$$
2x \cdot (10 - x) = \sqrt{x^2 + 100}
$$

$$
2x \cdot 10 - 2x^2 = \sqrt{x^2 + 100}
$$

Para facilitar a resolução, elevamos ambos os lados ao quadrado:

$$
\left(2x \cdot 10 - 2x^2\right)^2 = \left(\sqrt{x^2 + 100}\right)^2

# Questão 4 - Alternativa - Vale até 2 pontos

minimizar a área de superfície \( A \) de uma lata de sopa cilíndrica para um dado volume \( V \). A fórmula da área é dada por:

$$A = (\frac 2\pi rh + 8r^2)$$

onde \( r \) é o raio da base do cilindro e \( h \) é a altura. Como o volume \( V \) do cilindro é fixo e dado por:

$$V = (pi/r^2 h)$$

podemos isolar \( h \) em termos de \( r \) e \( V \):

$$h = (\frac{V}{\pi r^2})$$

Substituindo \( h \) na fórmula da área \( A \):

$$A = 2\pi r \left(\frac{V}{\pi r^2}\right) + 8r^2$$
$$= \frac{2V}{r} + 8r^2$$

Para encontrar o valor de \( r \) que minimiza \( A \), diferenciamos \( A \) em relação a \( r \) e igualamos a zero:

\frac{dA}{dr} = -\frac{2V}{r^2} + 16r = 0

Resolvendo para \( r \):

$$16r^3 = 2V$$

$$r^3 = \frac{V}{8}$$

$$r = \left(\frac{V}{8}\right)^{\frac{1}{3}}$$
 Usando a relação entre \( h \) e \( r \), encontramos \( h \):

$$h = \frac{V}{\pi r^2}$$

$$= \frac{V}{\pi \left(\left(\frac{V}{8}\right)^{\frac{1}{3}}\right)^2}$$

$$= \frac{V}{\pi \left(\frac{V^{\frac{2}{3}}}{64^{\frac{1}{3}}}\right)}$$

$$= \frac{8V^{\frac{1}{3}}}{\pi V^{\frac{2}{3}}}$$

$$= \frac{8}{\pi V^{\frac{1}{3}}}$$

Finalmente, a proporção entre a altura \( h \) e o diâmetro \( 2r \) é:

$$\frac{h}{2r} = \frac{\frac{8}{\pi V^{\frac{1}{3}}}}{2\left(\frac{V}{8}\right)^{\frac{1}{3}}}$$

$$= \frac{8}{\pi V^{\frac{1}{3}}} \times \frac{1}{2V^{\frac{1}{3}}}$$

$$= \frac{4}{\pi}$$

Portanto, a proporção entre a altura e o diâmetro para a lata mais econômica é $$\frac{4}{\pi}$$




## Solução Questão 5

$${Derivada Parcial em relação a \( x \) (\( F_x \)):}$$

Para encontrar a derivada parcial de \( F \) em relação a \( x \), denotada por \( F_x \), tratamos \( y \) como uma constante e derivamos em relação a \( x \):


\begin{align*}
\frac{\partial}{\partial x} x^3 & = 3x^2, \\
\frac{\partial}{\partial x} (3x^2y) & = 6xy, \\
\frac{\partial}{\partial x} (-2xy^2) & = -2y^2, \\
\frac{\partial}{\partial x} y^3 & = 0 \quad (\text{pois } y^3 \text{ é constante em relação a } x), \\
\frac{\partial}{\partial x} \cos(x + 2y) & = -\sin(x + 2y) \quad (\text{usando a regra da cadeia}).
\end{align*}


Somando todas essas derivadas parciais, temos:

F_x(x, y) = 3x^2 + 6xy - 2y^2 - \sin(x + 2y).


$$Derivada Parcial em relação a \( y \) (\( F_y \)):$$

Para encontrar a derivada parcial de \( F \) em relação a \( y \), denotada por \( F_y \), tratamos \( x \) como uma constante e derivamos em relação a \( y \):


\begin{align*}
\frac{\partial}{\partial y} x^3 & = 0 \quad (\text{pois } x^3 \text{ é constante em relação a } y), \\
\frac{\partial}{\partial y} (3x^2y) & = 3x^2, \\
\frac{\partial}{\partial y} (-2xy^2) & = -4xy, \\
\frac{\partial}{\partial y} y^3 & = 3y^2, \\
\frac{\partial}{\partial y} \cos(x + 2y) & = -2\sin(x + 2y) \quad (\text{usando a regra da cadeia}).
\end{align*}


Somando todas essas derivadas parciais, temos:

$$F_y(x, y) = 3x^2 - 4xy + 3y^2 - 2\sin(x + 2y).$$
